```python
# Обрати всі домогосподарства, у яких загальна активна споживана потужність перевищує 5 кВт.
df_active_power = df_cleaned[df_cleaned['Global_active_power'] > 5.0]
# Вивести результат
print("Домогосподарства з потужністю більше 5 кВт:")
print(df_active_power.loc[:, ['Date', 'Time', 'Global_active_power']])
```

```python
# Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.
df_voltage = df_cleaned[df_cleaned['Voltage'] > 235]
# Вивести результат
print("Домогосподарства, у яких вольтаж перевищую 235 В:")
print(df_voltage.loc[:, ['Date', 'Time', 'Voltage']])
```

```python
# Обрати домогосподарства зі силою струму в межах 19-20 А
df_selected_current = df_cleaned[(df_cleaned['Global_intensity'] >= 19) & (df_cleaned['Global_intensity'] <= 20)]

# Визначити ті, у яких пральна машина та холодильник 'Sub_metering_2' споживають більше, ніж бойлер та кондиціонер 'Sub_metering_3'
df_selected_appliances = df_selected_current[(df_selected_current['Sub_metering_2'] >  df_selected_current['Sub_metering_3'])]

# Вивести результат в одному рядку
print("Домогосподарства зі силою струму в межах 19-20 А та більшим споживанням пральної машини та холодильника:")
print(df_selected_appliances.loc[:, ['Date', 'Time', 'Global_intensity', 'Sub_metering_2', 'Sub_metering_3']].to_string(index=False))
```

```python
# Випадковим чином вибрати 500000 домогосподарств
random_sample = np.random.choice(df_cleaned.index, size=500000, replace=False)
df_random_sample = df_cleaned.loc[random_sample]

# Обчислити середні значення для трьох груп споживання електричної енергії
average_sub_metering_1 = df_random_sample['Sub_metering_1'].mean()
average_sub_metering_2 = df_random_sample['Sub_metering_2'].mean()
average_sub_metering_3 = df_random_sample['Sub_metering_3'].mean()

# Вивести результат
print("Середня споживана енергія (Sub_metering_1):", average_sub_metering_1)
print("Середня споживана енергія (Sub_metering_2):", average_sub_metering_2)
print("Середня споживана енергія (Sub_metering_3):", average_sub_metering_3)
```

```python
# Вибрати домогосподарства, які після 18:00 споживають понад 6 кВт за хвилину в середньому
df_selected_power = df_cleaned[(df_cleaned['Time'] > '18:00') & (df_cleaned['Global_active_power'] > 6.0)]

# Визначити ті, у яких основне споживання електроенергії припадає на пральну машину, сушарку, холодильник та освітлення
df_selected_group2 = df_selected_power[(df_selected_power['Sub_metering_2'] > df_selected_power[['Sub_metering_1', 'Sub_metering_3']].max(axis=1))]
#print(df_selected_group2[['Date', 'Time', 'Global_active_power', 'Sub_metering_2']].to_string(index=False))

# Порахувати кількість рядків
num_rows = df_selected_group2.shape[0]
# Вивести результат
#print(f"Кількість рядків: {num_rows}")

# Розділіть список на дві половини
mid_point = len(df_selected_group2) // 2
first_half = df_selected_group2[:mid_point]
second_half = df_selected_group2[mid_point:]

# Оберіть кожен третій елемент з першої половини та кожний четвертий елемент з другої половини
selected_elements = pd.concat([first_half.iloc[2::3], second_half.iloc[3::4]])

# Виведіть результат
print(selected_elements[['Date', 'Time', 'Global_active_power', 'Sub_metering_2']].to_string(index=False))

# Порахувати кількість рядків
num_rows2 = selected_elements.shape[0]
# Вивести результат
#print(f"Кількість рядків: {num_rows2}")
```
